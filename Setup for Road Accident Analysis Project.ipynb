# Setup for Road Accident Analysis Project

Data are available 
- https://www.data.gouv.fr/datasets/bases-de-donnees-annuelles-des-accidents-corporels-de-la-circulation-routiere-annees-de-2005-a-2024
- https://www.data.gouv.fr/datasets/referentiel-geographique-francais-communes-unites-urbaines-aires-urbaines-departements-academies-regions

```python
# Set default catalog
catalog = "training"
spark.sql(f"USE CATALOG {catalog}")
```

```python
# Create schemas
spark.sql(f"CREATE SCHEMA IF NOT EXISTS {catalog}.roaa_dev_raw") # Bronze layer
spark.sql(f"CREATE SCHEMA IF NOT EXISTS {catalog}.roaa_dev_dataprep") # Silver layer
spark.sql(f"CREATE SCHEMA IF NOT EXISTS {catalog}.roaa_dev_analytics") # Gold layer
```

```python
# Create volumes
spark.sql(f"CREATE VOLUME IF NOT EXISTS {catalog}.roaa_dev_raw.geographic_reference_system")

spark.sql(f"CREATE VOLUME IF NOT EXISTS {catalog}.roaa_dev_raw.baac_vehicle_registration")
spark.sql(f"CREATE VOLUME IF NOT EXISTS {catalog}.roaa_dev_raw.accident_characteristics")
spark.sql(f"CREATE VOLUME IF NOT EXISTS {catalog}.roaa_dev_raw.accident_sites")
spark.sql(f"CREATE VOLUME IF NOT EXISTS {catalog}.roaa_dev_raw.accident_victims")
spark.sql(f"CREATE VOLUME IF NOT EXISTS {catalog}.roaa_dev_raw.accident_vehicles")
```
